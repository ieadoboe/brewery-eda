# SPC Analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "bottle_dataset.csv"

ds = pd.read_csv(path)
```

```python
ds = ds.drop(columns=["Unnamed: 0"])
ds.tail()
```

```python
ds["Batch_ID"] = range(1, len(ds) + 1)

ds.head()
```

```python
np.random.seed(78)

# Generate new features
ds["Defective_Count"] = np.random.randint(0, 11, size=len(ds))
ds["Defect_Count"] = np.random.randint(6, 20, size=len(ds))
ds["Total_Produced"] = np.random.randint(50, 81, size=len(ds))

ds.head()
```

```python
ds.columns
```

```python
ds = ds[
    [
        "Batch_ID",
        "Beer_Style",
        "Location",
        "Fermentation_Time",
        "Temperature",
        "pH_Level",
        "Gravity",
        "Alcohol_Content",
        "Bitterness",
        "Volume_Produced",
        "Total_Sales",
        "Quality_Score",
        "Brewhouse_Efficiency",
        "Loss_During_Brewing",
        "Loss_During_Fermentation",
        "Loss_During_Bottling_Kegging",
        "Defective_Count",
        "Defect_Count",
        "Total_Produced",
    ]
]
ds.tail()
```

```python
ds.head()
```

```python
ds = ds[:100]
```

```python
ds.describe()
```

```python
# Calculate X-bar (Mean of Quality Score per Batch)
ds["X_Bar"] = ds["Quality_Score"]

# Compute control limits
mean_xbar = ds["X_Bar"].mean()
std_xbar = ds["X_Bar"].std()
UCL = mean_xbar + 3 * std_xbar  # Upper Control Limit (UCL)
LCL = mean_xbar - 3 * std_xbar  # Lower Control Limit (LCL)

# Plot X-bar Chart
plt.figure(figsize=(12, 6))
plt.plot(ds["Batch_ID"], ds["X_Bar"], marker="o", linestyle="-", label="X (Batch Mean)")
plt.axhline(mean_xbar, color="blue", linestyle="dashed", label="Process Mean")
plt.axhline(UCL, color="red", linestyle="dashed", label="UCL (+3)")
plt.axhline(LCL, color="red", linestyle="dashed", label="LCL (-3)")

# Highlight out-of-control points
out_of_control = ds[(ds["X_Bar"] > UCL) | (ds["X_Bar"] < LCL)]
plt.scatter(
    out_of_control["Batch_ID"],
    out_of_control["X_Bar"],
    color="red",
    zorder=3,
    label="Out of Control",
)

# Labels & Title
plt.xlabel("Batch ID")
plt.ylabel("Quality Score")
plt.title("X-bar Control Chart")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate Range per Batch (Max - Min within each batch)
range_per_batch = (
    ds.groupby("Beer_Style")["Quality_Score"]
    .agg(lambda x: x.max() - x.min())
    .reset_index()
)
range_per_batch.rename(columns={"Quality_Score": "Range"}, inplace=True)

# Compute Mean Range (R)
R_bar = range_per_batch["Range"].mean()

# Control Chart Constants for n=5 samples per batch
D3, D4 = 0, 2.114  # Standard values from SPC tables

# Compute Control Limits
UCL_R = D4 * R_bar
LCL_R = D3 * R_bar

# Plot R-chart
plt.figure(figsize=(12, 6))
plt.plot(
    range_per_batch["Beer_Style"],
    range_per_batch["Range"],
    marker="o",
    linestyle="-",
    label="Range per Batch",
)
plt.axhline(R_bar, color="blue", linestyle="dashed", label="Mean Range (R)")
plt.axhline(UCL_R, color="red", linestyle="dashed", label="UCL (D4 * R)")
plt.axhline(LCL_R, color="red", linestyle="dashed", label="LCL (D3 * R)")

# Highlight out-of-control points
out_of_control_r = range_per_batch[range_per_batch["Range"] > UCL_R]
plt.scatter(
    out_of_control_r["Beer_Style"],
    out_of_control_r["Range"],
    color="red",
    zorder=3,
    label="Out of Control",
)

# Labels & Title
plt.xlabel("Beer_Style")
plt.ylabel("Range (Max-Min within Batch)")
plt.title("R-Chart (Range Control Chart)")
plt.ylim(0, UCL_R * 1.1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
samples_per_batch = 200 / 8


# Calculate Proportion Defective per Batch (p_i)
ds["Proportion_Defective"] = ds["Defective_Count"] / ds["Total_Produced"]

# Compute Overall Proportion Defective (p)
p_bar = ds["Proportion_Defective"].mean()

# Compute Control Limits
UCL_p = p_bar + 3 * np.sqrt((p_bar * (1 - p_bar)) / samples_per_batch)
LCL_p = max(
    0, p_bar - 3 * np.sqrt((p_bar * (1 - p_bar)) / samples_per_batch)
)  # LCL cannot be negative

# Plot P-Chart
plt.figure(figsize=(12, 6))
plt.plot(
    ds["Batch_ID"],
    ds["Proportion_Defective"],
    marker="o",
    linestyle="-",
    label="Proportion Defective",
)
plt.axhline(p_bar, color="blue", linestyle="dashed", label="Mean Proportion (p)")
plt.axhline(UCL_p, color="red", linestyle="dashed", label="UCL (+3)")
plt.axhline(LCL_p, color="red", linestyle="dashed", label="LCL (-3, min 0)")

# Highlight out-of-control points
out_of_control_p = ds[
    (ds["Proportion_Defective"] > UCL_p) | (ds["Proportion_Defective"] < LCL_p)
]
plt.scatter(
    out_of_control_p["Batch_ID"],
    out_of_control_p["Proportion_Defective"],
    color="red",
    zorder=3,
    label="Out of Control",
)

# Labels & Title
plt.xlabel("Batch ID")
plt.ylabel("Proportion Defective")
plt.title("P-Chart (Proportion Control Chart)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute Mean Defective Count (np)
np_bar = ds["Defective_Count"].mean()

# Compute Control Limits
UCL_np = np_bar + 3 * np.sqrt(np_bar * (1 - (np_bar / samples_per_batch)))
LCL_np = max(
    0, np_bar - 3 * np.sqrt(np_bar * (1 - (np_bar / samples_per_batch)))
)  # LCL cannot be negative

# Plot NP-Chart
plt.figure(figsize=(12, 6))
plt.plot(
    ds["Batch_ID"],
    ds["Defective_Count"],
    marker="o",
    linestyle="-",
    label="Defective Count",
)
plt.axhline(np_bar, color="blue", linestyle="dashed", label="Mean Defective Count (np)")
plt.axhline(UCL_np, color="red", linestyle="dashed", label="UCL (+3)")
plt.axhline(LCL_np, color="red", linestyle="dashed", label="LCL (-3, min 0)")

# Highlight out-of-control points
out_of_control_np = ds[
    (ds["Defective_Count"] > UCL_np) | (ds["Defective_Count"] < LCL_np)
]
plt.scatter(
    out_of_control_np["Batch_ID"],
    out_of_control_np["Defective_Count"],
    color="red",
    zorder=3,
    label="Out of Control",
)

# Labels & Title
plt.xlabel("Batch ID")
plt.ylabel("Number of Defective Units")
plt.title("NP-Chart (Number Defective Control Chart)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute Defects per Unit (u_i) for each batch
ds["Defects_Per_Unit"] = ds["Defect_Count"] / ds["Total_Produced"]

# Compute Mean Defects per Unit ()
U_bar = ds["Defects_Per_Unit"].mean()

# Compute batch-specific control limits
ds["UCL"] = U_bar + 3 * np.sqrt(U_bar / ds["Total_Produced"])
ds["LCL"] = np.maximum(0, U_bar - 3 * np.sqrt(U_bar / ds["Total_Produced"]))

# Plot U-Chart
plt.figure(figsize=(12, 6))
plt.plot(
    ds["Batch_ID"],
    ds["Defects_Per_Unit"],
    marker="o",
    linestyle="-",
    label="Defects Per Unit (u)",
)
plt.plot(ds["Batch_ID"], ds["UCL"], linestyle="dashed", color="red", label="UCL (+3)")
plt.plot(ds["Batch_ID"], ds["LCL"], linestyle="dashed", color="red", label="LCL (-3)")
plt.axhline(U_bar, color="blue", linestyle="dashed", label="Mean ")

# Highlight out-of-control points
out_of_control_u = ds[
    (ds["Defects_Per_Unit"] > ds["UCL"]) | (ds["Defects_Per_Unit"] < ds["LCL"])
]
plt.scatter(
    out_of_control_u["Batch_ID"],
    out_of_control_u["Defects_Per_Unit"],
    color="red",
    zorder=3,
    s=80,  # Slightly larger points for better visibility
    label="Out of Control",
)

# Labels & Title
plt.xlabel("Batch ID")
plt.ylabel("Defects Per Unit")
plt.title("U-Chart (Defects Per Unit Control Chart)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## C-chart


```python
# Compute Mean Defect Count (C)
C_bar = ds["Defect_Count"].mean()

# Compute Control Limits
UCL_c = C_bar + 3 * np.sqrt(C_bar)
LCL_c = max(0, C_bar - 3 * np.sqrt(C_bar))  # LCL cannot be negative

# Plot C-Chart
plt.figure(figsize=(12, 6))
plt.plot(
    ds["Batch_ID"], ds["Defect_Count"], marker="o", linestyle="-", label="Defect Count"
)
plt.axhline(C_bar, color="blue", linestyle="dashed", label="Mean Defect Count (C)")
plt.axhline(UCL_c, color="red", linestyle="dashed", label="UCL (+3)")
plt.axhline(LCL_c, color="red", linestyle="dashed", label="LCL (-3, min 0)")

# Highlight out-of-control points
out_of_control_c = ds[(ds["Defect_Count"] > UCL_c) | (ds["Defect_Count"] < LCL_c)]
plt.scatter(
    out_of_control_c["Batch_ID"],
    out_of_control_c["Defect_Count"],
    color="red",
    zorder=3,
    label="Out of Control",
)

# Labels & Title
plt.xlabel("Batch ID")
plt.ylabel("Number of Defects")
plt.title("C-Chart (Defect Count Control Chart)")
plt.legend()
plt.grid(True)
plt.show()
```
